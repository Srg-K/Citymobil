```python
ops = {
    "+": (lambda x,y: x+y),
    "-": (lambda x,y: x-y),
    "*": (lambda x,y: x*y),
    "//": (lambda x,y: x//y),
    "%": (lambda x,y: x%y),
    "**": (lambda x,y: x**y)
}
```

```python
res = []
for line in data:
    res. append(ops[line.split('    ')[0]](int(line.split('    ')[1]), int(line.split('    ')[2])))
```

```python
f1 = open('import_file_2_1.txt', 'r')
f2 = open('import_file_2_2.txt', 'r')

data1 = f1.readlines()
data2 = f2.readlines()

res = ''
for i in range(len(data1)):
    s1 = data1[i]
    s2 = data2[i]
    
    cursor = [sum(x) for x in zip(*[[int(x) for x in s2.split(' ')], [0, 1]])]
    res += s1[cursor[0]:cursor[1]] + ' '
    
res = res[0:len(res)-1]
```
