```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('titanic.csv')
```

```python
data
```

```python
male, female = int(data[data['Sex']=='male']['Sex'].value_counts()), int(data[data['Sex']=='female']['Sex'].value_counts())
print(male, female)
```

```python
survived = data[['Survived','PassengerId']].groupby('Survived').count().reset_index()
survived['share'] = np.divide(survived['PassengerId'], survived['PassengerId'].sum())
survived
```

```python
fc = data[['Pclass','PassengerId']].groupby('Pclass').count().reset_index()
fc['share'] = np.divide(fc['PassengerId'], fc['PassengerId'].sum())
fc
```

```python
print(data['Age'].mean(), int(data['Age'].median()))
```

```python
data[['SibSp','Parch']].corr(method='pearson')
```

```python
w_names = data[data['Sex']=='female'][['Name','PassengerId']]
w_names['first_name'] = w_names['Name'].apply(lambda x: x.split(',')[0])

w_names = w_names[['first_name','PassengerId']].groupby('first_name').count().reset_index()
w_names = w_names.sort_values(by='PassengerId', ascending=False)
w_names[:1]
```
